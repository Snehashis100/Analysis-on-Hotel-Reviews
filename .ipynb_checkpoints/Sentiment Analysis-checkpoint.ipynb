```python
import pandas as pd
```

```python
from nltk.corpus import stopwords
```

```python
data=pd.read_csv("sentiment_data.csv")
data.head(3)
```

```python
data=data.drop('Unnamed: 0',axis=1)
```

```python
data.isnull().sum()
```

```python
data=data[data['Reviewer_Nationality'].notna()]
data.isnull().sum()
```

```python
# Convert the reviews to lower and delete leading/trailing space
data["Negative_Review"] = data["Negative_Review"].str.lower().str.strip()
data["Positive_Review"] = data["Positive_Review"].str.lower().str.strip()
```

```python
import nltk
nltk.download()
```

```python
# SentimentIntensityAnalyzer is a pretrained model on reviews which help us to find the positivity and negativity in a statement
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sent_analyzer = SentimentIntensityAnalyzer()

rev1 = "The hotel was very good, I love it!"
rev2 = "It was just horrible, the worst ever."

print(f"review 1:\n{rev1}\nScore: {sent_analyzer.polarity_scores(rev1)}")

print(f"\nreview 2:\n{rev2}\nScore: {sent_analyzer.polarity_scores(rev2)}")
```

```python
# the compound key is the overall positivity or negativity of a review it will be greater than 0 if the review is positive 
# and less than zero if it is negative
```

```python
# Take only a part of the data to speed up
import time
df = data[:100000].copy()

start_time = time.time()
pos = df["Positive_Review"].apply(lambda x: abs(sent_analyzer.polarity_scores(x)["compound"]))
neg = df["Negative_Review"].apply(lambda x: -abs(sent_analyzer.polarity_scores(x)["compound"]))

df["sentiment_score"] = pos + neg
df["polarity_pos"] = pos
df["polarity_neg"] = neg

time_model = time.time() - start_time
print(f"Execution time: {int(time_model)} seconds")
```

```python
negative_reviews=[]
data=data.reset_index(drop=True)
for i in range(len(data)):
    if data.loc[i,'Negative_Review'].lower().strip()!='no negative':
        negative_reviews.append(data.loc[i,'Negative_Review'])
```

```python
len(negative_reviews)
```

```python
# Taking the 1st 10000 records for faster computation
negative_reviews=negative_reviews[:10000]
```

```python
len(negative_reviews)
```

```python
negative_reviews[0]
```

```python
word_token=[]
for sent in negative_reviews:
    word_token.append(sent.split())
```

```python
word_token[0]
```

```python
for i in range(len(word_token)):
    word_token[i] = [word for word in word_token[i] if word not in stopwords.words('english')]
```

```python
word_token[0]
# After removing the stopwords
```

```python
# converting the words into vectors to see the semantic relationship among the words
import nltk
from gensim.models import Word2Vec
```

```python
model = Word2Vec(word_token,min_count=2)
```

```python
angry_idx = model.wv.key_to_index["angry"]
```

```python
angry_cnt = model.wv.get_vecattr("angry", "count")
```
